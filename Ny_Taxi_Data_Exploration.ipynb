```python
import pandas as pd

# Convert to datetime (if not already)
df["lpep_pickup_datetime"] = pd.to_datetime(df["lpep_pickup_datetime"])

# Filter trips with distance < 100
filtered_df = df[df["trip_distance"] < 100]

# Get row with maximum trip distance
max_trip_row = filtered_df.loc[filtered_df["trip_distance"].idxmax()]

# Extract values
max_distance = max_trip_row["trip_distance"]
pickup_day = max_trip_row["lpep_pickup_datetime"].date()

print("Max trip distance:", max_distance)
print("Pickup day:", pickup_day)

```

```python
import pandas as pd

zones = pd.read_csv("taxi_zone_lookup.csv")
zones.head()
```

```python
nov18_df = df[
    (df["lpep_pickup_datetime"] >= "2025-11-18") &
    (df["lpep_pickup_datetime"] < "2025-11-19")
]
```

```python
pickup_totals = (
    nov18_df
    .groupby("PULocationID")["total_amount"]
    .sum()
    .reset_index()
)
```

```python
pickup_with_zones = pickup_totals.merge(
    zones,
    left_on="PULocationID",
    right_on="LocationID",
    how="left"
)
```

```python
top_zone = pickup_with_zones.loc[
    pickup_with_zones["total_amount"].idxmax()
]

top_zone[["Zone", "Borough", "total_amount"]]
```

```python
df["lpep_pickup_datetime"] = pd.to_datetime(df["lpep_pickup_datetime"])
```

```python
ehn_id = zones.loc[zones["Zone"] == "East Harlem North", "LocationID"].iloc[0]

ehn_id
```

```python
nov_ehn_df = df[
    (df["lpep_pickup_datetime"] >= "2025-11-01") &
    (df["lpep_pickup_datetime"] < "2025-12-01") &
    (df["PULocationID"] == ehn_id)
]
```

```python
tip_by_dropoff = (
    nov_ehn_df
    .groupby("DOLocationID")["tip_amount"]
    .sum()
    .reset_index()
)
```

```python
tip_with_zones = tip_by_dropoff.merge(
    zones,
    left_on="DOLocationID",
    right_on="LocationID",
    how="left"
)
```

```python
top_dropoff = tip_with_zones.loc[
    tip_with_zones["tip_amount"].idxmax()
]

top_dropoff[["Zone", "Borough", "tip_amount"]]
```
